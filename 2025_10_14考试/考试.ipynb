## 1.以下程序的功能是对随机产生的 1～100 之间的一个整数 n，利用二分法查找这个数并输出查找的次数。


```python
import random
import math

n = random.randint(1, 100)  # 函数 random.randint(a,b) 返回从 a 到 b 之间随机选出的一个整数
a, b, cnt = 1, 100, 0
while a <= b:
    cnt += 1
    m = math.ceil((a + b) / 2)  # 函数 math.ceil(x) 的返回值为数值 x 的向上取整
    if n < m:
        b = m - 1
    elif n > m:
        a = m + 1
    else:
        print(m)
        break
print(cnt)
```

## 2.以下程序读入正整数 n，计算
 S = 1! − 2! + 3! − 4! + 5! − …… ± n!
 并输出。

```python
n = int(input())
S = 0
sign = 1
k = 0
for i in range(1, n + 1):
    k = 1
    for j in range(i):
        k *= (j + 1)
    S += sign * k
    sign = -sign
print(S)
```

## 3. 下面程序中的 Change 函数的功能是将十进制数转化 N 进制数（2 <= N <= 16）。
 输入目标进制和待转换的数，函数返回转换后的目标进制数。


```python
def Change(radix, num):
    g = []
    while num != 0:
        r = num % radix
        num = num // radix
        g.append(r)
    ans = ''
    t = ('A', 'B', 'C', 'D', 'E', 'F')
    for x in range(len(g)):
        if g[x] >= 10:
            ans += t[g[x] - 10]
        else:
            ans += str(g[x])
    return ans[::-1]


print(Change(16, 12455))
```

## 4. 给定一个整数的数字列表 L，判断 L 中是否存在相同数字，若存在返回 True，否则返回 False。
 请在横线上补全代码。


```python
def identical(L):
    flag = 0
    L.sort()
    for i in range(len(L) - 1):
        if L[i] == L[i + 1]:
            flag = 1
            break
    if flag:
        return True
    else:
        return False
```

## 5. 程序填空。一个列表 a 中记录了连续多天的营业额，求其中最长的**连续增长（不含相等）**的子序列的长度。
 例如 [5, 1, 2, 3, 5, 9, 6, 7, 4, 6, 8] 中最长的连续增长（不含相等）的子序列长度是 5（即 [1, 2, 3, 5, 9] 的长度）。

```python
a = [5, 1, 2, 3, 5, 9, 6, 7, 4, 6, 8]
maxlen = 1
n = 1
for i in range(1, len(a)):
    if a[i - 1] < a[i]:
        n += 1
        if maxlen < n:
            maxlen = n
    else:
        n = 1
print(maxlen)
```

# python深度学习


- 1.简述 Batch Normalization 的基本原理与作用。
- Batch Normalization(批量归一化)
基本原理:在训练过程中对每个批次的输入数据进行归一化处理，将输入数据进行中心化和缩放，使得每个批次的输入数据分布更接近标准正太分布
通常操作:对输入数据进行均值和方差计算，并减去均值，再除以方差,将结果乘以一个缩放因子，并加上一个偏置项
作用:批量归一化可以加速训练，并提高模型的泛化性能
- 2.什么是 Dropout？它在深度神经网络训练中的作用是什么？
- Dropout(随机失活)
在每个训练步骤中，每个神经元都有暂时被“删除”的概率。经过dropout训练的神经元不能与其相邻的神经元相互适应，也不能过分依赖少数输入神经元，它们必须注意每个输入神经元。它们最终对输入的微小变化不太敏感。
作用:获得一个更有鲁棒性的网络，该网络有更好的泛化能力。
- 3.解释梯度裁剪的概念及其在训练过程中的作用。
- 梯度裁剪
在反向传播过程中裁剪梯度，使它们永远不会超过某个阈值。
作用:它可以防止梯度爆炸，从而提高训练的稳定性。
- 4.什么是早停？为什么它能防止过拟合？
- 早停
早停是机器学习的一种策略，它通过在训练过程中观察模型在验证集上的表现，并停止训练，当模型在验证集上的表现不再改善时，就停止训练，防止过拟合。
- 5.数据增强（Data Augmentation）在训练过程中的意义是什么？请举例说明。
- 数据增强
它可以通过改变原始数据来生成新的数据，从而提高模型的泛化能力。
举例：改变图片大小,位置,方向
- 6.什么是梯度消失？ 为什么会出现？ 有哪些解决办法？
- 梯度消失
当神经元数量过多时，梯度可能会因为神经元数量过多而消失。
解决办法：使用Batch Normalization(批量归一化)，使用ReLU激活函数，使用Dropout(随机失活)，使用梯度裁剪等
- 7.什么是学习率调度？有什么好处？
- 学习率调度
它可以通过改变学习率来提高模型的训练效果。
好处：可以提高模型的训练效果,避免过拟合

# 深度学习预测序列


```python
import numpy as np
import tensorflow as tf

# 假设我们用 60 天的价格预测第 61 天价格
time_steps = 60
features = 1
samples = 200

# 构造输入数据 (samples, time_steps, features)
x_train = np.random.rand(samples, time_steps, features)  # 200个 60天股票价格
y_train = np.random.rand(samples, 1)  # 200个 对应的第61天股票价格
x_test = np.random.rand(10, time_steps, features)  # 测试数据
stock_model = tf.keras.Sequential()
stock_model.add(tf.keras.layers.LSTM(units=4, input_shape=(time_steps, features), return_sequences=True))
stock_model.add(tf.keras.layers.Dropout(rate=0.2))
stock_model.add(tf.keras.layers.LSTM(units=4, return_sequences=False))
stock_model.add(tf.keras.layers.Dropout(rate=0.2))
stock_model.add(tf.keras.layers.Dense(units=1))
stock_model.compile(loss='mean_squared_error', optimizer='adam')
stock_model.fit(x_train, y_train, epochs=5, batch_size=16, verbose=0)
stock_model.summary()
predictions = stock_model.predict(x_test)
print(predictions[:, -1])
```
